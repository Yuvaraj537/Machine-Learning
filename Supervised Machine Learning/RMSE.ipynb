# Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Assuming 'df' is your DataFrame with 'Expenditure', 'Savings', and 'Salary' columns

# Prepare the data
X = df[['Expenditure', 'Savings']]
y = df['Salary']

# Fit the regression model
model = LinearRegression()
model.fit(X, y)

# Create a figure for 3D plotting
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of actual data points
ax.scatter(df['Expenditure'], df['Savings'], df['Salary'], color='blue', label="Actual Salary")

# Create a grid for the surface plot
x_surf, y_surf = np.meshgrid(np.linspace(df['Expenditure'].min(), df['Expenditure'].max(), 10),
                             np.linspace(df['Savings'].min(), df['Savings'].max(), 10))

# Predict the z values for the regression plane
z_surf = model.intercept_ + model.coef_[0] * x_surf + model.coef_[1] * y_surf

# Plot the surface
ax.plot_surface(x_surf, y_surf, z_surf, alpha=0.5, color='red')

# Set labels and title
ax.set_xlabel("Expenditure")
ax.set_ylabel("Savings")
ax.set_zlabel("Salary")
ax.set_title("3D Visualization of Multiple Linear Regression")

# Show the plot
plt.show()
```

# MSE

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
rmse = np.sqrt(mse)

print('MSE : ', mse)
print('MAE : ', mae)
print('RMSE : ', rmse)
```
